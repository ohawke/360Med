```python
directory = r'C:\Users\owenh\Documents'
dict_list = []
data = pd.read_csv("export.csv")
with open(directory + '\\export.dic', 'r') as f: 
    lines = f.readlines()
    curr_dict = {}
    good_var = 0 
    for line in lines: 
        if (line.startswith('[Format=')):
            curr_dict = {}
            good_var = 1
            continue
        if (good_var == 1 and line != "\n"):
            print(line)
            key, value = line.strip().split("=") 
            key = int(key)
            curr_dict[key] = value 
        if (good_var == 1 and line == "\n"):
            dict_list.append(curr_dict)
dict_list[2]
```

```python
data["SEER registry"] = data["SEER registry"].map(dict_list[0])
data["Cause of death recode"] = data["Cause of death recode"].map(dict_list[1])
data["Year of death"] = data["Year of death"].map(dict_list[2])
data
```

```python
@interact
def do_search(Search='tumor'):
    cpt_query = {'string':Search, 'apiKey':apikey, 'sabs': 'CPT', 'returnIdType':'code'}
    output = requests.get(base_url+path, params=cpt_query)
    output.encoding = 'utf-8'
    outputJson = output.json()
    results = (([outputJson['result']])[0])['results']
    for r in results:
        if r['ui'] in cpt_codes.index:
            print(r['name'])
            print(cpt_codes.loc[r['ui']][5:7])
    clin_query = {'expr':Search, 'fmt':'JSON'}
    response = requests.get(clin_url, params=clin_query)
    print(response.json())
```
